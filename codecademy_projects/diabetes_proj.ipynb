### <span style='color:HotPink'>Exploratory data analysis project on diabetes patient information</span>

```python
# imports
import pandas as pd
import numpy as np
import os
```

```python
# set path to dataset
input_folder = '/Users/alexandrabrown/Desktop/data_science/datasets/EDA_Diagnosing_Diabetes'
input_file   = 'diabetes.csv'
input_path   = os.path.join(input_folder, input_file)

df = pd.read_csv(input_path)
df
```

```python
# are there any missing values?
df.info()
```

```python
# compute summary stats to further explore if values are missing/abnormal
df.describe()
# even though there are not technically any missing values some values are 0, which would not make sense for a BMI
# or for some of the other columns
```

```python
# replace 0 values with NaN
cols = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']

df[cols] = df[cols].replace(0, np.nan)

# now recheck missing entries
df.info()
```

```python
# do the rows with missing data have something in common? view these rows
df[df.isnull().any(axis=1)]
```

```python
# look at data types of the cols
df.dtypes
```

```python
# why is the Outcome column object (meaning string type ?)
df['Outcome'].unique()
# looks like theres 'o' where there should be '0'
```

```python
# replace the 'O' with '0'
df['Outcome'] = df['Outcome'].replace('O', 0)
# then set as type int
df['Outcome'] = df['Outcome'].astype('int')
# double check the data types again
df.dtypes
```
