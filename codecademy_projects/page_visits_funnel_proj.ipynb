```python
# imports
import pandas as pd
import numpy as np
import os
```

```python
# set paths to data
input_folder  = '/Users/alexandrabrown/Desktop/data_science/datasets/page_visits_funnel'
cart_file     = 'cart.csv'
checkout_file = 'checkout.csv'
purchase_file = 'purchase.csv'
visits_file   = 'visits.csv'

cart_path     = os.path.join(input_folder, cart_file)
checkout_path = os.path.join(input_folder, checkout_file)
purchase_path = os.path.join(input_folder, purchase_file)
visits_path   = os.path.join(input_folder, visits_file)

# load the data
cart     = pd.read_csv(cart_path)
checkout = pd.read_csv(checkout_path)
purchase = pd.read_csv(purchase_path)
visits   = pd.read_csv(visits_path)
```

### <span style='color:violet'>2) Combine visits & cart using left merge</span>

```python
visits_cart = pd.merge(visits, cart, how='left')
visits_cart
```

### <span style='color:violet'>4) How many entries in cart_time are NaN?<span>

```python
count_na = visits_cart[visits_cart['cart_time'].isna() == True].reset_index(drop=True)
print(f'The number of NaN entries in cart_time is: {len(count_na)}')
print('These NaN entries represent instances where someone visited the website but did not place anything in their cart')
```

### <span style='color:violet'>5) What percentage of people who visited the website ended up not placing anything in their cart?</span>

```python
all_visits = float(len(visits_cart))
no_cart    = float(len(count_na))

perc_no_cart = (no_cart / all_visits)*100
print(f'The percentage of people who visited but did not place anything in their cart is: {perc_no_cart}%')
```

### <span style='color:violet'>6) Left merge for cart and checkout</span>
What percentage of users put items in their cart, but did not proceed to checkout?

```python
cart_checkout = pd.merge(cart, checkout, how='left')

# get number of people that put something in cart
cart_count = float(len(cart_checkout))
# get number of people that checked out
checkout_na = cart_checkout[cart_checkout['checkout_time'].isna()]
checkout_na_count = float(len(checkout_na))

# get percent of people that did not checkout
perc_no_checkout = (checkout_na_count / cart_count)*100
print(f'The percentage of users that put items in cart but did not checkout is: {perc_no_checkout:.2f}%')
```

### <span style='color:violet'>7) Left merge all 4 DataFrames of the funnel in order</span>

```python
all_data = visits.merge(cart, how='left').merge(checkout, how='left').merge(purchase, how='left')
all_data
```

### <span style='color:violet'>10) Time from initial visit to final purchase</span>
add new column to `all_data` to show this difference

```python
# first convert the time columns (str) to date time
all_data['purchase_time'] = pd.to_datetime(all_data['purchase_time'])
all_data['visit_time'] = pd.to_datetime(all_data['visit_time'])


all_data['time_to_purchase'] = all_data['purchase_time'] - all_data['visit_time']
all_data
```

### <span style='color:violet'>12) Compute average time to purchase from the new column</span>

```python
avg_time = all_data['time_to_purchase'].mean()
print(f'The average time to purchase is: {avg_time}')
```
